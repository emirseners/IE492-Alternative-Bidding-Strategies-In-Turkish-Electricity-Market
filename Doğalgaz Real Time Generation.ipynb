```python
import pandas as pd
from matplotlib import pyplot as plt
ADA_files = ["ADA Gercek_Zamanli_Uretim-01012024-31032024.csv", 
             "ADA Gercek_Zamanli_Uretim-01042024-30062024.csv", 
             "ADA Gercek_Zamanli_Uretim-01072024-30092024.csv"]

dfs = [pd.read_csv(file, sep=";", parse_dates=["Tarih"]) for file in ADA_files]
ADA = pd.concat(dfs, ignore_index=True)
ADA["Datetime"] = pd.to_datetime(ADA["Tarih"] + " " + ADA["Saat"], dayfirst=True)
ADA = ADA.set_index("Datetime")
ADA = ADA[["Doğal Gaz"]]


HAMİTABAT_files = ["HAMİTABAT Gercek_Zamanli_Uretim-01012024-31032024.csv", 
                   "HAMİTABAT Gercek_Zamanli_Uretim-01042024-30062024.csv", 
                   "HAMİTABAT Gercek_Zamanli_Uretim-01072024-30092024.csv"]

dfs = [pd.read_csv(file, sep=";", parse_dates=["Tarih"]) for file in HAMİTABAT_files]
HAMİTABAT = pd.concat(dfs, ignore_index=True)
HAMİTABAT["Datetime"] = pd.to_datetime(HAMİTABAT["Tarih"] + " " + HAMİTABAT["Saat"], dayfirst=True)
HAMİTABAT = HAMİTABAT.set_index("Datetime")
HAMİTABAT = HAMİTABAT[["Doğal Gaz"]]


KIRIKKALE_files = ["KIRIKKALE Gercek_Zamanli_Uretim-01012024-31032024.csv", 
                   "KIRIKKALE Gercek_Zamanli_Uretim-01042024-30062024.csv", 
                   "KIRIKKALE Gercek_Zamanli_Uretim-01072024-30092024.csv"]

dfs = [pd.read_csv(file, sep=";", parse_dates=["Tarih"]) for file in KIRIKKALE_files]
KIRIKKALE = pd.concat(dfs, ignore_index=True)
KIRIKKALE["Datetime"] = pd.to_datetime(KIRIKKALE["Tarih"] + " " + KIRIKKALE["Saat"], dayfirst=True)
KIRIKKALE = KIRIKKALE.set_index("Datetime")
KIRIKKALE = KIRIKKALE[["Doğal Gaz"]]


RWE_DENİZLİ_files = ["RWE DENİZLİ Gercek_Zamanli_Uretim-01012024-31032024.csv", 
                     "RWE DENİZLİ Gercek_Zamanli_Uretim-01042024-30062024.csv", 
                     "RWE DENİZLİ Gercek_Zamanli_Uretim-01072024-30092024.csv"]

dfs = [pd.read_csv(file, sep=";", parse_dates=["Tarih"]) for file in RWE_DENİZLİ_files]
RWE_DENİZLİ = pd.concat(dfs, ignore_index=True)
RWE_DENİZLİ["Datetime"] = pd.to_datetime(RWE_DENİZLİ["Tarih"] + " " + RWE_DENİZLİ["Saat"], dayfirst=True)
RWE_DENİZLİ = RWE_DENİZLİ.set_index("Datetime")
RWE_DENİZLİ = RWE_DENİZLİ[["Doğal Gaz"]]


TEKIRDAG_files = ["TEKİRDAĞ Gercek_Zamanli_Uretim-01012024-31032024.csv", 
                  "TEKİRDAĞ Gercek_Zamanli_Uretim-01042024-30062024.csv", 
                  "TEKİRDAĞ Gercek_Zamanli_Uretim-01072024-30092024.csv"]

dfs = [pd.read_csv(file, sep=";", parse_dates=["Tarih"]) for file in TEKIRDAG_files]
TEKIRDAG = pd.concat(dfs, ignore_index=True)
TEKIRDAG["Datetime"] = pd.to_datetime(TEKIRDAG["Tarih"] + " " + TEKIRDAG["Saat"], dayfirst=True)
TEKIRDAG = TEKIRDAG.set_index("Datetime")
TEKIRDAG = TEKIRDAG[["Doğal Gaz"]]


YENI_ELEKTRIK_files = ["YENİ ELEKTRİK Gercek_Zamanli_Uretim-01012024-31032024.csv", 
                       "YENİ ELEKTRİK Gercek_Zamanli_Uretim-01042024-30062024.csv", 
                       "YENİ ELEKTRİK Gercek_Zamanli_Uretim-01072024-30092024.csv"]

dfs = [pd.read_csv(file, sep=";", parse_dates=["Tarih"]) for file in YENI_ELEKTRIK_files]
YENI_ELEKTRIK = pd.concat(dfs, ignore_index=True)
YENI_ELEKTRIK["Datetime"] = pd.to_datetime(YENI_ELEKTRIK["Tarih"] + " " + YENI_ELEKTRIK["Saat"], dayfirst=True)
YENI_ELEKTRIK = YENI_ELEKTRIK.set_index("Datetime")
YENI_ELEKTRIK = YENI_ELEKTRIK[["Doğal Gaz"]]
```

```python
dfs = [ADA, HAMİTABAT, KIRIKKALE, RWE_DENİZLİ, TEKIRDAG, YENI_ELEKTRIK]
df = pd.concat(dfs, axis=1)
df.columns = ["ADA", "HAMİTABAT", "KIRIKKALE", "RWE_DENİZLİ", "TEKİRDAĞ", "YENİ_ELEKTRİK"]
df = df.replace({r',': ''}, regex=True).apply(pd.to_numeric)
df
```
