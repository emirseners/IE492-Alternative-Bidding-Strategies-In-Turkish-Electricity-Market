```python
import pandas as pd
from matplotlib import pyplot as plt

ZETES1_files = ["ZETES1 Gercek_Zamanli_Uretim-01012024-31032024.csv", 
                "ZETES1 Gercek_Zamanli_Uretim-01042024-30062024.csv", 
                "ZETES1 Gercek_Zamanli_Uretim-01072024-30092024.csv"]

dfs = [pd.read_csv(file, sep=";", parse_dates=["Tarih"]) for file in ZETES1_files]
ZETES1 = pd.concat(dfs, ignore_index=True)
ZETES1["Datetime"] = pd.to_datetime(ZETES1["Tarih"] + " " + ZETES1["Saat"], dayfirst=True)
ZETES1 = ZETES1.set_index("Datetime")
ZETES1 = ZETES1[["İthal Kömür"]]


ZETES2_files = ["ZETES2 Gercek_Zamanli_Uretim-01012024-31032024.csv", 
                "ZETES2 Gercek_Zamanli_Uretim-01042024-30062024.csv", 
                "ZETES2 Gercek_Zamanli_Uretim-01072024-30092024.csv"]

dfs = [pd.read_csv(file, sep=";", parse_dates=["Tarih"]) for file in ZETES2_files]
ZETES2 = pd.concat(dfs, ignore_index=True)
ZETES2["Datetime"] = pd.to_datetime(ZETES2["Tarih"] + " " + ZETES2["Saat"], dayfirst=True)
ZETES2 = ZETES2.set_index("Datetime")
ZETES2 = ZETES2[["İthal Kömür"]]


ZETES3_files = ["ZETES3 Gercek_Zamanli_Uretim-01012024-31032024.csv", 
                "ZETES3 Gercek_Zamanli_Uretim-01042024-30062024.csv", 
                "ZETES3 Gercek_Zamanli_Uretim-01072024-30092024.csv"]

dfs = [pd.read_csv(file, sep=";", parse_dates=["Tarih"]) for file in ZETES3_files]
ZETES3 = pd.concat(dfs, ignore_index=True)
ZETES3["Datetime"] = pd.to_datetime(ZETES3["Tarih"] + " " + ZETES3["Saat"], dayfirst=True)
ZETES3 = ZETES3.set_index("Datetime")
ZETES3 = ZETES3[["İthal Kömür"]]


İSKEN_files = ["İSKEN Real-Time_Generation-01012024-31032024.csv", 
               "İSKEN Real-Time_Generation-01042024-30062024.csv", 
               "İSKEN Real-Time_Generation-01072024-30092024.csv"]

dfs = [pd.read_csv(file, sep=";", parse_dates=["Date"]) for file in İSKEN_files]
İSKEN = pd.concat(dfs, ignore_index=True)
İSKEN["Datetime"] = pd.to_datetime(İSKEN["Date"] + " " + İSKEN["Hour"], dayfirst=True)
İSKEN = İSKEN.set_index("Datetime")
İSKEN = İSKEN[["Import Coal"]]


HUNUTLU_files = ["HUNUTLU Real-Time_Generation-01012024-31032024.csv", 
                 "HUNUTLU Real-Time_Generation-01042024-30062024.csv", 
                 "HUNUTLU Real-Time_Generation-01072024-30092024.csv"]

dfs = [pd.read_csv(file, sep=";", parse_dates=["Date"]) for file in HUNUTLU_files]
HUNUTLU = pd.concat(dfs, ignore_index=True)
HUNUTLU["Datetime"] = pd.to_datetime(HUNUTLU["Date"] + " " + HUNUTLU["Hour"], dayfirst=True)
HUNUTLU = HUNUTLU.set_index("Datetime")
HUNUTLU = HUNUTLU[["Import Coal"]]


CENAL_files = ["CENAL Real-Time_Generation-01012024-31032024.csv", 
               "CENAL Real-Time_Generation-01042024-30062024.csv", 
               "CENAL Real-Time_Generation-01072024-30092024.csv"]

dfs = [pd.read_csv(file, sep=";", parse_dates=["Date"]) for file in CENAL_files]
CENAL = pd.concat(dfs, ignore_index=True)
CENAL["Datetime"] = pd.to_datetime(CENAL["Date"] + " " + CENAL["Hour"], dayfirst=True)
CENAL = CENAL.set_index("Datetime")
CENAL = CENAL[["Import Coal"]]


ATLAS_files = ["ATLAS Real-Time_Generation-01012024-31032024.csv", 
               "ATLAS Real-Time_Generation-01042024-30062024.csv", 
               "ATLAS Real-Time_Generation-01072024-30092024.csv"]

dfs = [pd.read_csv(file, sep=";", parse_dates=["Date"]) for file in ATLAS_files]
ATLAS = pd.concat(dfs, ignore_index=True)
ATLAS["Datetime"] = pd.to_datetime(ATLAS["Date"] + " " + ATLAS["Hour"], dayfirst=True)
ATLAS = ATLAS.set_index("Datetime")
ATLAS = ATLAS[["Import Coal"]]


AFŞİN_ELBİSTAN_files = ["AFŞİN ELBİSTAN Real-Time_Generation-01012024-31032024.csv", 
                        "AFŞİN ELBİSTAN Real-Time_Generation-01042024-30062024.csv", 
                        "AFŞİN ELBİSTAN Real-Time_Generation-01072024-30092024.csv"]

dfs = [pd.read_csv(file, sep=";", parse_dates=["Date"]) for file in AFŞİN_ELBİSTAN_files]
AFŞİN_ELBİSTAN = pd.concat(dfs, ignore_index=True)
AFŞİN_ELBİSTAN["Datetime"] = pd.to_datetime(AFŞİN_ELBİSTAN["Date"] + " " + AFŞİN_ELBİSTAN["Hour"], dayfirst=True)
AFŞİN_ELBİSTAN = AFŞİN_ELBİSTAN.set_index("Datetime")
AFŞİN_ELBİSTAN = AFŞİN_ELBİSTAN[["Lignite"]]
```

```python
dfs = [ZETES1, ZETES3, ATLAS, HUNUTLU, ZETES2, CENAL, İSKEN]
df = pd.concat(dfs, axis=1)
df.columns = ["ZETES1", "ZETES3", "ATLAS", "HUNUTLU", "ZETES2", "CENAL", "İSKEN"]
df = df.replace({r',': ''}, regex=True).apply(pd.to_numeric)
df
```
